```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train1 = np.array([1,2,3])
y_train1 = np.array([1,2,3])
m1 = len(x_train1)

plt.scatter(x_train1,y_train1, color="tab:red", marker="X")
```

### Cost function for linear regression:

![Alt text](images/costfunction.png)

```python
def generate_prediction_vector(x,m,w,b):
   y_wb = w * x + b # "x" is a vector
   return y_wb

def calculate_cost_vector(x,y,m,wb): # Calculates the cost
    J_wb = np.zeros(len(wb))
    for i in range(len(wb)):  
      J_wb[i] = 1/(2*m) * sum((generate_prediction_vector(x_train1,m1,wb[i][0],wb[i][1])) - y)**2
    return J_wb
```

```python
w = np.arange(0.5,1.5,0.01)
# b = np.arange(0,10,0.1)
b = np.zeros(len(w))
wb = np.array([w,b]).transpose() # Pairs of w and b

J_wb = calculate_cost_vector(x_train1,y_train1,m1,wb)

plt.figure()
plt.plot(w,J_wb)
plt.title("Cost function")
plt.xlabel("W")
plt.ylabel("$J_{wb}(w,0)$")
plt.xlim(0.5,1.5)
plt.grid()
```
